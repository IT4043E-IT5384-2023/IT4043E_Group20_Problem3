```python
from os import listdir
from os.path import join
import json
import pandas as pd
import datetime
```

```python
my_path='xdata_raw'
names = listdir(my_path)
```

```python
df = pd.DataFrame({
    'id' : pd.Series(dtype='str'),
    'username' : pd.Series(dtype='str'),
    'favourites_count': pd.Series(dtype='float'),
    'followers_count': pd.Series(dtype='float'),
    'friends_count' : pd.Series(dtype='float'),
    'protected': pd.Series(dtype='float'),
    'verified' : pd.Series(dtype='float'),
    'tweet_frequency' : pd.Series(dtype='float'),
    'retweet_rate': pd.Series(dtype='float'),
    'sensitive_rate': pd.Series(dtype='float'),
    'mean_reply_count' : pd.Series(dtype='float'),
    'mean_quote_count' : pd.Series(dtype='float'),
    'mean_bookmark_count' : pd.Series(dtype='float'),
    'mean_views' : pd.Series(dtype='float'),
    'mean_likes': pd.Series(dtype='float'),
    'mean_retweet_count' : pd.Series(dtype='float'),
})
```

```python
for i, user in enumerate(names):
    files = listdir(join(my_path, user))

    print('[{}/{}] {}'.format(i + 1, len(names), user))

    if(len(files)==0): 
        continue

    fi = open(join(my_path,user,files[0]), 'r', encoding='utf8')
    f = json.loads(fi.read())

    id = str(f['author']['id'])
    username = str(f['author']['username'])
    favourites_count = float(f['author']['favourites_count']) if f['author']['favourites_count'] else 0.0
    followers_count = float(f['author']['followers_count']) if f['author']['followers_count'] else 0.0
    friends_count = float(f['author']['friends_count']) if f['author']['friends_count'] else 0.0

    protected = 0.0
    if(f['author']['protected']):
        protected = 1.0
    verified = 0.0
    if(f['author']['verified']):
        verified = 1.0
    start = datetime.datetime.strptime(f['created_on'], '%Y-%m-%d %H:%M:%S%z')
    end = datetime.datetime.strptime(f['created_on'], '%Y-%m-%d %H:%M:%S%z')
    tweet_frequency = 0.0
    retweet_rate = 0.0
    sensitive_rate = 0.0
    mean_reply_count = 0.0
    mean_quote_count = 0.0
    mean_bookmark_count = 0.0
    mean_views = 0.0
    mean_likes = 0.0
    mean_retweet_count = 0.0

    fi.close()

    for file in files:
        fi = open(join(my_path,user,file), 'r', encoding='utf8')
        f = json.loads(fi.read())
        
        start = min(start, datetime.datetime.strptime(f['created_on'], '%Y-%m-%d %H:%M:%S%z'))
        end = max(start, datetime.datetime.strptime(f['created_on'], '%Y-%m-%d %H:%M:%S%z'))
        if(f['is_retweet']):
            retweet_rate += 1.0
        if(f['is_sensitive']):
            sensitive_rate += 1.0
        mean_reply_count += float(f['reply_counts'])
        mean_quote_count += float(f['quote_counts'])
        mean_bookmark_count += float(f['bookmark_count'])
        if isinstance(f['views'], int):
            mean_views += float(f['views'])
        mean_likes += float(f['likes'])
        mean_retweet_count += float(f['retweet_counts'])

        fi.close()

    tweet_frequency = len(files) / (float((end - start).days) + 1.0)
    retweet_rate = retweet_rate / len(files)
    sensitive_rate = sensitive_rate / len(files)
    mean_reply_count = mean_reply_count / len(files)
    mean_quote_count = mean_quote_count / len(files)
    mean_bookmark_count = mean_bookmark_count / len(files)
    mean_likes = mean_likes / len(files)
    mean_retweet_count = mean_retweet_count / len(files)
    mean_views = mean_views / len(files)

    df.loc[len(df)] = [
        id,
        username,
        favourites_count,
        followers_count, 
        friends_count,
        protected,
        verified,
        tweet_frequency,
        retweet_rate,
        sensitive_rate,
        mean_reply_count,
        mean_quote_count,
        mean_bookmark_count,
        mean_views,
        mean_likes,
        mean_retweet_count
    ]

df.to_csv('x_processed.csv')
```
